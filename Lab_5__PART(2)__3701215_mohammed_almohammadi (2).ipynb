```python
# Importing required libraries.
import pandas as pd
import numpy as np 
import seaborn as sns #visualsation
import matplotlib.pyplot as plt #visualsation
%matplotlib inline 
sns.set(color_codes=True)
```

```python
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
```

```python
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), '*')
plt.plot(x, np.cos(x), '--')
```

```python
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.tan(x), '--')
```
